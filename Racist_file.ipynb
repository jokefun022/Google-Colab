```python
import re
import pandas as pd
from pathlib import Path

# Set your file paths (in Colab, upload your file first or use drive mount)
INPUT_PATH = Path("1 Racism Excel File.xlsx")  # Replace with your file name
OUTPUT_PATH = Path("1_Racism_Excel_File_with_emojis.xlsx")

# Load workbook (first sheet)
df = pd.read_excel(INPUT_PATH, sheet_name=0)

# Choose primary text column: object dtype and most non-null entries
text_cols = [c for c in df.columns if df[c].dtype == object or pd.api.types.is_string_dtype(df[c])]
if not text_cols:
    text_col = df.columns[0]
else:
    text_col = max(text_cols, key=lambda c: df[c].notna().sum())

# Keyword-to-emoji map
keyword_map = {
    r"\bidiot\b": ["ü§°","üí•"],
    r"\bstupid\b": ["ü§Ø","ü§¶‚Äç‚ôÇÔ∏è"],
    r"\bfuck\b": ["ü§¨","üí¢"],
    r"\bshit\b": ["üí©","ü§Æ"],
    r"\bbhag\w*\b": ["üèÉ‚Äç‚ôÇÔ∏è","üí®"],
    r"\bmaderch\w*\b": ["ü§¨","üí¢"],
    r"\bmadarjhat\b": ["ü§¨","üí¢"],
    r"\bch\w+ya\b": ["ü§¨","üí¢"],
    r"\bmaro\b": ["üî™","üí•"],
    r"\bkafir\b": ["ü§¨","üí¢"],
    r"\bkaffir\b": ["ü§¨","üí¢"],
    r"\bnigga\b": ["üí¢","ü§¨"],
    r"\bnigger\b": ["üí¢","ü§¨"],
    r"\bwasooli\b": ["üòí","üí¢"],
    r"\btatti\b": ["üí©","ü§Æ"],
    r"\bclown\b": ["ü§°","üòÇ"],
    r"\bpathetic\b": ["üòí","ü§¶"],
    r"\btrash\b": ["üóëÔ∏è","üí©"],
    r"\bscum\b": ["ü§¢","üí©"],
    r"\bsick\b": ["ü§¢","ü§Æ"],
    r"\bkill\b": ["üî™","üí•"],
    r"\bdie\b": ["üíÄ","üòµ"],
    r"\bslut\b": ["üòí","ü§¶‚Äç‚ôÄÔ∏è"],
}

compiled_keywords = [(re.compile(k, flags=re.IGNORECASE), v) for k, v in keyword_map.items()]

def place_emojis(sentence):
    if not isinstance(sentence, str) or sentence.strip() == "":
        return sentence

    s = sentence
    emojis_added = []
    offset = 0
    matches_found = 0

    for pattern, emojis in compiled_keywords:
        for m in pattern.finditer(s):
            if matches_found >= 2:
                break
            emoji_to_insert = emojis[0] if emojis else "ü§¨"
            s = s[:m.end()+offset] + emoji_to_insert + s[m.end()+offset:]
            offset += len(emoji_to_insert)
            emojis_added.append(emoji_to_insert)
            matches_found += 1
            break
        if matches_found >= 2:
            break

    needed = max(0, 2 - len(emojis_added))
    if needed > 0:
        if len(emojis_added) >= 1:
            mid = len(s) // 2
            pos = s.find(" ", mid)
            if pos == -1:
                pos = len(s)
            emoji_choice = "ü§°" if "ü§¨" in emojis_added or "ü§Ø" in emojis_added else "üí•"
            s = s[:pos] + emoji_choice + s[pos:]
            emojis_added.append(emoji_choice)
            needed -= 1
    if needed > 0:
        l = len(s)
        positions = [max(1, int(l*0.3)), max(1, int(l*0.7))]
        for pos in positions[:needed]:
            p = s.find(" ", pos)
            if p == -1:
                p = len(s)
            emoji_choice = "üí¢" if "ü§¨" in emojis_added else "üòí"
            s = s[:p] + emoji_choice + s[p:]
            emojis_added.append(emoji_choice)

    if sum(1 for ch in s if ch in "ü§¨ü§Øü§°üí•üí©ü§Æüí¢üòíü§¶üî™üíÄüòµüòÇüóëÔ∏èü§¢ü§¶‚Äç‚ôÄÔ∏èü§¶‚Äç‚ôÇÔ∏è") < 2:
        s = s + "ü§¨üí•"

    return s

# Apply
new_col = text_col + "_with_emojis"
df[new_col] = df[text_col].apply(place_emojis)

# Save output
df.to_excel(OUTPUT_PATH, index=False)

print(f"‚úÖ File saved as: {OUTPUT_PATH}")
```
